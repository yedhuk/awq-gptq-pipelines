```python
from awq import AutoAWQForCausalLM
from transformers import AutoTokenizer
```

```python

# 1. Define paths and configuration
model_path = 'meta-llama/Llama-3.2-3B-Instruct'
quant_path = './models/llama-3.2-3b-instruct-awq-custom' # Saves to a local directory

```

# AWQ config

```python
# Standard AWQ configuration for a 4-bit model
quant_config = { 
    "zero_point": True, 
    "q_group_size": 128, 
    "w_bit": 4, 
    "version": "GEMM" 
}

```

# Load FP16 model

```python
print(f"Loading base model from: {model_path}")
# 2. Load the base model and tokenizer into memory
model = AutoAWQForCausalLM.from_pretrained(model_path)
tokenizer = AutoTokenizer.from_pretrained(model_path, trust_remote_code=True)

```

# Quantization

```python
print("Starting the quantization process (this may take a few minutes)...")
# 3. Quantize the model
model.quantize(tokenizer, quant_config=quant_config)

```

# Save Quantized Tokenizer and Model

```python
print(f"Saving quantized weights to local folder: {quant_path}")
# 4. Save the new quantized model and tokenizer
model.save_quantized(quant_path)
tokenizer.save_pretrained(quant_path)

print("Done! You can now load this model directly from the local folder.")
```
